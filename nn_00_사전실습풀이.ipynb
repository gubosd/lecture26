```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. 아래 그림과 같은 샘플 데이터를 생성하고 직접 그래프를 그리시오.
    - 샘플 갯수는 100개
    - 샘플들은 y = 2*X + 1 직선 근방으로 분포
    - 랜덤값 활용 (np.random.randn() 함수 등)
    - 데이터 변수이름은 각각 X 와 y (점들의 X 좌표값과 y 좌표값에 해당함)
    


2. 입력층과 출력층만 가지는 신경망을 설계하고 훈련하시오.
    - X 가 입력, y 가 목표값 (입력 속성이 1개임)
    - 선형회귀에 해당함
    - 손실값은 MSE(Mean Squared Error) 적용
<br/><br/>

3. 훈련 결과를 아래와 같이 표시하시오.
    - 손실값(loss) 감소 그래프 그리기
    - 최종 가중치 값 출력 (기울기가 2, 절편값이 1 과 비슷한지 확인)
    - 예측 모델이 직선임을 확인

```python
X = np.random.randn(100)
y = 2*X + 1 + np.random.randn(100)*0.5
```

```python
plt.scatter(X, y, alpha=0.5)

xs = np.array([-3,3])
plt.plot(xs, xs*2+1, 'r:', label='y = 2*X + 1')

plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,)))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
h = model.fit(X.reshape(-1,1), y, epochs=500, batch_size=32)
```

```python
plt.plot(h.history['loss'])
```

```python
ws = model.get_weights()
w = ws[0][0,0]; b=ws[1][0]
w, b
```

```python
xs = np.arange(-3, 3, 0.1)

y_pred = model.predict(xs.reshape(-1,1)).ravel()
y_pred.shape
```

```python
plt.scatter(X, y, alpha=0.5)
plt.plot(xs, y_pred, 'r:', label=f'w={w:.5}\nb={b:.5}')
plt.legend()
```
